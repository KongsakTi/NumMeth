3) How many **non-zero** terms do we need to calculate $\cos(0.1)$ such that the bound on the **truncation error** is less than $10^{-7}$

```python
x = 0.1
def counting(x):
    term = 1
    bound  = x**term/math.factorial(term)
    while(bound > math.pow(10,-7)):
        bound = x**term/math.factorial(term)
        term += 1
    return term

print counting(x)
```

4) Plot $\ln (1+x)$ from $-0.5$ to $0.5$ along with the Taylor approximation using (1,2,3,4) terms. I want 5 lines on the same plot.

```python
xs = np.linspace(-0.5,0.5,100)
def one_term(x):
    return x
def two_term(x):
    return x - x**2/2
def three_term(x):
    return x - x**2/2 + x**3/3
def four_term(x):
    return x - x**2/2 + x**3/3 - x**4/4

ys = np.log1p(xs)
plt.plot(xs, ys, color = 'black')
plt.plot(xs,one_term(xs),color='blue')
plt.plot(xs,two_term(xs),color='red')
plt.plot(xs,three_term(xs),color='green')
plt.plot(xs,four_term(xs),color='yellow')
```

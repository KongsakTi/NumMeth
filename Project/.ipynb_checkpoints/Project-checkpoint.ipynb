```python
def toBW(pic):
    w, h, d = pic.shape
    BW = np.zeros(pic.shape)
    
    for i in xrange(w):
        for j in xrange(h):
            if (sum(pic[i,j][:-1]**2) < 2.6):
#             if (sum(pic[i,j][:-1]) < 0.5):
                BW[i,j] = [1,1,1,0]
            else:
                BW[i,j] = [0,0,0,1]
    return BW

def convert(direct, filename):
    pic = plt.imread(direct + filename)

    BW = toBW(pic)
    
    plt.figure(figsize=(20,10))
    plt.imshow(BW)
    plt.savefig("BW_" + filename)

    return pic, BW
```
